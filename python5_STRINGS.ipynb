# some code for practice

```python
rows = int(input("enter the number of rows"))
for i in range(1,rows+1):
    print("*"*i)
```

```python
rows = int(input("enter the number of rows"))
for i in range(1,rows+1):
  print("*"*i,end=" ")
  for j in range(1,rows+1):
    print("")
```

```python
rows = int(input("enter the number of rows"))
for i in range(1,rows+1):
  for j in range(1,i+1):
    print("*",end="  ")
  print("")
```
