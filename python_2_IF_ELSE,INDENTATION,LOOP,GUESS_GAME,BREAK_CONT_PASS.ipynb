# Guessing Game

```python
import random
```

```python
random.randint(1,100)
```

```python
from typing import Counter
jackpot = random.randint(1,100)
guess = int(input("Guess the number"))
Counter=1
while guess != jackpot:
    if guess < jackpot:
      print('GUESS HIGHER')
    else:
      print('GUESS LOWER')
    guess = int(input("Guess the number"))
    Counter+=1
print('CORRECT GUESS')
print("YOU TOOK", Counter,"ATTEMPTS")
```

# Nested loops

```python
rows=int(input("enter the number of rows"))
for i in range(1,rows+1):
  for j in range(1,i+1):
    print("*",end=" ")
  print("")

```

# same code execution


```python
rows = int(input("Enter the number of rows"))
for i in range(1,rows+1):
  print("* "*i)
```

# break-continue-and-pass

```python
for i in range(1,11):
  if i==5:
     break
  print(i)
```

```python
# break is used in LINEAR SEARCHING
```

```python
# continue is used in skip something
# like in e-commerce website we have to display the product on the frontpage which are in stock
# so by doing loop using ("if") statement to check whether the item is in stock or not
# if it is in stock then display if non in stock then continue
```

```python
for i in range(1,11):
  if i == 6:
    continue
  print(i)
```

```python
for i in range(1,11):
  if i == 4:
    continue
  print(i)
  print("hello")
```
